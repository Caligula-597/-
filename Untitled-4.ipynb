```python
from __future__ import annotations

# Experiment 6 (v2): Empirical anchoring for a single 3D flock event
# Single-file notebook script
#
# Main upgrades over v1:
#   1. Extended topological-rank analysis (default max_rank=100)
#   2. Smoothed rank-correlation curve
#   3. More stable cumulative-mass effective-rank proxies:
#        k_eff_50, k_eff_75, k_eff_90
#   4. Smoothed zero-cross estimate if available
#
# Interpretation remains conservative:
#   - empirical anchor
#   - strong collective order / correlation regime
#   - low-degree compatibility can be discussed only if proxies actually support it
#
# Not claimed:
#   - exact empirical entropy production
#   - direct proof of J = dot(S)/chi minimization in nature

import json
import re
from dataclasses import dataclass
from pathlib import Path
from typing import Any, Dict, Optional, Sequence, Tuple

import jax
import jax.numpy as jnp
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt


# -----------------------------
# 0. Global configuration
# -----------------------------

jax.config.update("jax_enable_x64", True)


@dataclass
class EmpiricalConfig:
    input_path: str = "sd01.rtf"
    outdir: str = "./exp6_empirical_anchor_v2"

    delta_t: float = 1.0

    # Metric correlation binning
    n_r_bins: int = 48
    r_max_quantile: float = 0.95
    min_pairs_per_bin: int = 50

    # Topological rank correlation
    max_rank: int = 100
    rank_smooth_window: int = 5   # odd integer recommended

    # Empirical local mismatch k-scan
    k_grid: Tuple[int, ...] = (3, 4, 5, 6, 7, 9, 12, 15, 20)

    # Window language inherited from experiments 4/5
    core_window_k: Tuple[int, ...] = (3, 4, 5)
    shoulder_k: Tuple[int, ...] = (6, 7)

    # Optional model summary for regime overlay
    model_window_csv: Optional[str] = None

    save_plots: bool = True
    dpi: int = 180


# -----------------------------
# 0.5 Utility helpers
# -----------------------------

def trapz_compat(y, x):
    if hasattr(np, "trapezoid"):
        return np.trapezoid(y, x)
    return np.trapz(y, x)


def moving_average_1d(x: np.ndarray, window: int) -> np.ndarray:
    if window <= 1:
        return x.copy()
    if window % 2 == 0:
        window += 1
    pad = window // 2
    xpad = np.pad(x, (pad, pad), mode="edge")
    kernel = np.ones(window, dtype=np.float64) / window
    return np.convolve(xpad, kernel, mode="valid")


def first_zero_crossing_linear(x: np.ndarray, y: np.ndarray) -> float | None:
    """Return interpolated x at first y-crossing from positive to <= 0."""
    valid = np.isfinite(x) & np.isfinite(y)
    xv = x[valid]
    yv = y[valid]
    if len(xv) < 2:
        return None
    for i in range(len(xv) - 1):
        if yv[i] > 0 and yv[i + 1] <= 0:
            x0, x1 = xv[i], xv[i + 1]
            y0, y1 = yv[i], yv[i + 1]
            if abs(y1 - y0) < 1e-15:
                return float(x1)
            return float(x0 - y0 * (x1 - x0) / (y1 - y0))
    return None


# -----------------------------
# 1. File parsing
# -----------------------------

def parse_sd01_ascii_table(path: str | Path) -> pd.DataFrame:
    """
    Robust parser for the starling event file.

    Supports:
      - plain text files
      - RTF files containing numeric rows with optional trailing backslashes

    Expected numeric columns:
      x(t1), y(t1), z(t1), x(t2), y(t2), z(t2)
    """
    path = Path(path)
    if not path.exists():
        raise FileNotFoundError(f"Input file not found: {path}")

    text = path.read_text(encoding="utf-8", errors="ignore")

    header_key = "#  x(t1)"
    start = text.find(header_key)
    if start == -1:
        raise ValueError(
            "Could not find the numeric table header '#  x(t1) ... z(t2)' in the file."
        )

    body = text[start:]

    num = r"[-+]?(?:\d*\.\d+|\d+)"
    row_pattern = re.compile(
        rf"({num})\s+({num})\s+({num})\s+({num})\s+({num})\s+({num})(?:\\)?"
    )
    rows = row_pattern.findall(body)

    if not rows:
        raise ValueError(
            "No numeric 6-column rows found after the data header. "
            "The file format may differ from the expected event table."
        )

    return pd.DataFrame(
        [[float(x) for x in row] for row in rows],
        columns=["x1", "y1", "z1", "x2", "y2", "z2"],
    )


# -----------------------------
# 2. Core geometry / kinematics
# -----------------------------

def normalize_rows(x: jnp.ndarray, eps: float = 1e-12) -> jnp.ndarray:
    norms = jnp.linalg.norm(x, axis=1, keepdims=True)
    norms = jnp.maximum(norms, eps)
    return x / norms


def pairwise_distances(x: jnp.ndarray) -> jnp.ndarray:
    diff = x[:, None, :] - x[None, :, :]
    d2 = jnp.sum(diff * diff, axis=-1)
    d2 = jnp.maximum(d2, 0.0)
    return jnp.sqrt(d2)


def compute_basic_state(df: pd.DataFrame, delta_t: float) -> Dict[str, Any]:
    r1 = jnp.asarray(df[["x1", "y1", "z1"]].to_numpy(np.float64))
    r2 = jnp.asarray(df[["x2", "y2", "z2"]].to_numpy(np.float64))

    v = (r2 - r1) / delta_t
    speed = jnp.linalg.norm(v, axis=1)
    u = normalize_rows(v)
    u_bar = jnp.mean(u, axis=0)
    delta_u = u - u_bar

    phi = float(jnp.linalg.norm(jnp.mean(u, axis=0)))
    mean_speed = float(jnp.mean(speed))

    return {
        "r1": r1,
        "r2": r2,
        "v": v,
        "speed": speed,
        "u": u,
        "u_bar": u_bar,
        "delta_u": delta_u,
        "phi": phi,
        "mean_speed": mean_speed,
    }


# -----------------------------
# 3. Connected metric correlation C(r)
# -----------------------------

def metric_connected_correlation(
    r: jnp.ndarray,
    delta_u: jnp.ndarray,
    n_bins: int,
    r_max_quantile: float,
    min_pairs_per_bin: int,
) -> Dict[str, Any]:
    """
    C(r) = <delta u_i . delta u_j>_{|r_ij| in bin} / <|delta u|^2>
    """
    n = r.shape[0]
    dmat = pairwise_distances(r)
    dots = delta_u @ delta_u.T

    iu = jnp.triu_indices(n, k=1)
    dvals = dmat[iu]
    cvals = dots[iu]

    denom = jnp.mean(jnp.sum(delta_u * delta_u, axis=1))
    denom = float(jnp.maximum(denom, 1e-15))

    dvals_np = np.asarray(dvals)
    cvals_np = np.asarray(cvals)

    r_max = float(np.quantile(dvals_np, r_max_quantile))
    edges = np.linspace(0.0, r_max, n_bins + 1)
    centers = 0.5 * (edges[:-1] + edges[1:])

    C = np.full(n_bins, np.nan, dtype=np.float64)
    counts = np.zeros(n_bins, dtype=np.int64)

    which = np.digitize(dvals_np, edges) - 1
    for b in range(n_bins):
        mask = which == b
        cnt = int(mask.sum())
        counts[b] = cnt
        if cnt >= min_pairs_per_bin:
            C[b] = cvals_np[mask].mean() / denom

    xi_0 = first_zero_crossing_linear(centers, C)
    if xi_0 is None:
        xi_0 = np.nan

    chi_C = np.nan
    xi_int = np.nan
    if np.isfinite(xi_0):
        mask = np.isfinite(C) & (centers <= xi_0)
        if mask.sum() >= 2:
            r_pos = centers[mask]
            c_pos = C[mask]
            if r_pos[-1] < xi_0:
                r_pos = np.concatenate([r_pos, [xi_0]])
                c_pos = np.concatenate([c_pos, [0.0]])
            xi_int = float(trapz_compat(c_pos, r_pos))
            chi_C = float(4.0 * np.pi * trapz_compat((r_pos ** 2) * c_pos, r_pos))

    return {
        "r_edges": edges,
        "r_centers": centers,
        "C_r": C,
        "counts": counts,
        "xi_0": xi_0,
        "xi_int": xi_int,
        "chi_C": chi_C,
        "denom": denom,
        "r_max": r_max,
    }


# -----------------------------
# 4. Topological-rank correlation C^(n)
# -----------------------------

def topological_rank_correlation(
    r: jnp.ndarray,
    delta_u: jnp.ndarray,
    max_rank: int,
    smooth_window: int,
) -> Dict[str, Any]:
    """
    Raw:
        C^(n) = <delta u_i . delta u_{j_n(i)}> / <|delta u|^2>

    Also returns:
        - smoothed curve
        - first zero-cross (raw and smoothed)
        - cumulative positive-mass effective ranks: 50%, 75%, 90%
    """
    n = r.shape[0]
    max_rank = min(max_rank, n - 1)

    dmat = np.array(pairwise_distances(r), copy=True)
    np.fill_diagonal(dmat, np.inf)
    order = np.argsort(dmat, axis=1)[:, :max_rank]

    du = np.asarray(delta_u)
    denom = float(max(np.mean(np.sum(du * du, axis=1)), 1e-15))

    Cn = np.zeros(max_rank, dtype=np.float64)
    for rank in range(max_rank):
        nbr = order[:, rank]
        dots = np.sum(du * du[nbr], axis=1)
        Cn[rank] = dots.mean() / denom

    ranks = np.arange(1, max_rank + 1)
    Cn_smooth = moving_average_1d(Cn, smooth_window)

    # zero-cross proxies
    k_eff_0_raw = first_zero_crossing_linear(ranks.astype(np.float64), Cn)
    k_eff_0_smooth = first_zero_crossing_linear(ranks.astype(np.float64), Cn_smooth)

    # cumulative positive mass
    pos_raw = np.maximum(Cn, 0.0)
    total_pos_raw = float(pos_raw.sum())

    def cumulative_rank(pos_arr: np.ndarray, frac: float) -> float | None:
        total = float(pos_arr.sum())
        if total <= 0:
            return None
        cs = np.cumsum(pos_arr)
        idx = np.searchsorted(cs, frac * total, side="left")
        idx = min(idx, len(ranks) - 1)
        return float(ranks[idx])

    k_eff_50 = cumulative_rank(pos_raw, 0.50)
    k_eff_75 = cumulative_rank(pos_raw, 0.75)
    k_eff_90 = cumulative_rank(pos_raw, 0.90)

    return {
        "ranks": ranks,
        "C_n": Cn,
        "C_n_smooth": Cn_smooth,
        "k_eff_0_raw": k_eff_0_raw,
        "k_eff_0_smooth": k_eff_0_smooth,
        "k_eff_50": k_eff_50,
        "k_eff_75": k_eff_75,
        "k_eff_90": k_eff_90,
        "neighbor_order": order,
    }


# -----------------------------
# 5. Empirical local-mismatch cost proxy
# -----------------------------

def empirical_local_mismatch_vs_k(
    u: jnp.ndarray,
    r: jnp.ndarray,
    k_grid: Sequence[int],
) -> pd.DataFrame:
    """
    c_mis(k) = <1 - u_i . Ahat_i^(k)>_i
    """
    n = r.shape[0]
    dmat = np.array(pairwise_distances(r), copy=True)
    np.fill_diagonal(dmat, np.inf)
    order = np.argsort(dmat, axis=1)

    u_np = np.asarray(u)

    rows = []
    for k in k_grid:
        if k >= n:
            continue
        nbr = order[:, :k]
        A = u_np[nbr].mean(axis=1)
        A_norm = np.linalg.norm(A, axis=1, keepdims=True)
        A_norm = np.maximum(A_norm, 1e-12)
        Ahat = A / A_norm
        cost_i = 1.0 - np.sum(u_np * Ahat, axis=1)
        rows.append({
            "k": int(k),
            "c_mis_mean": float(cost_i.mean()),
            "c_mis_median": float(np.median(cost_i)),
            "c_mis_std": float(cost_i.std(ddof=1)) if len(cost_i) > 1 else 0.0,
        })

    return pd.DataFrame(rows)


# -----------------------------
# 6. Empirical response-cost proxy J_emp_proxy(k)
# -----------------------------

def build_empirical_proxy_table(
    mismatch_df: pd.DataFrame,
    chi_C: float,
) -> pd.DataFrame:
    out = mismatch_df.copy()
    if chi_C is None or (not np.isfinite(chi_C)) or chi_C <= 0:
        out["J_emp_proxy"] = np.nan
    else:
        out["J_emp_proxy"] = out["c_mis_mean"] / chi_C
    return out


# -----------------------------
# 7. Optional regime mapping against model outputs
# -----------------------------

def maybe_load_csv(path: Optional[str]) -> Optional[pd.DataFrame]:
    if path is None:
        return None
    p = Path(path)
    if not p.exists():
        print(f"[warn] Optional file not found: {p}")
        return None
    return pd.read_csv(p)


def nearest_model_rows(
    model_df: pd.DataFrame,
    phi_emp: float,
    xi_emp: float,
    n_select: int = 5,
) -> Optional[pd.DataFrame]:
    if model_df is None:
        return None

    work = model_df.copy()
    phi_col = None
    xi_col = None
    for cand in ["phi_mean", "phi", "Phi"]:
        if cand in work.columns:
            phi_col = cand
            break
    for cand in ["xi", "xi_mean", "corr_length"]:
        if cand in work.columns:
            xi_col = cand
            break
    if phi_col is None or xi_col is None:
        return None

    work = work[np.isfinite(work[phi_col]) & np.isfinite(work[xi_col])].copy()
    if len(work) == 0:
        return None

    phi_scale = max(work[phi_col].std(ddof=0), 1e-12)
    xi_scale = max(work[xi_col].std(ddof=0), 1e-12)
    work["dist_emp"] = np.sqrt(
        ((work[phi_col] - phi_emp) / phi_scale) ** 2 +
        ((work[xi_col] - xi_emp) / xi_scale) ** 2
    )
    work = work.sort_values("dist_emp").head(n_select)
    return work


# -----------------------------
# 8. Plotting
# -----------------------------

def savefig(path: str | Path, dpi: int = 180) -> None:
    plt.tight_layout()
    plt.savefig(path, dpi=dpi, bbox_inches="tight")
    plt.close()


def plot_metric_correlation(metric: Dict[str, Any], outdir: Path, dpi: int) -> None:
    plt.figure(figsize=(6.4, 4.6))
    plt.plot(metric["r_centers"], metric["C_r"], marker="o", lw=1.5)
    plt.axhline(0.0, ls="--", lw=1.0)
    if np.isfinite(metric["xi_0"]):
        plt.axvline(metric["xi_0"], ls=":", lw=1.2, label=f"xi_0={metric['xi_0']:.2f}")
        plt.legend()
    plt.xlabel("metric distance r")
    plt.ylabel("connected velocity correlation C(r)")
    plt.title("Empirical connected metric correlation")
    savefig(outdir / "metric_correlation.png", dpi=dpi)


def plot_rank_correlation(rankcorr: Dict[str, Any], outdir: Path, dpi: int) -> None:
    plt.figure(figsize=(7.0, 4.8))
    plt.plot(rankcorr["ranks"], rankcorr["C_n"], marker="o", ms=3, lw=1.0, alpha=0.7, label="raw")
    plt.plot(rankcorr["ranks"], rankcorr["C_n_smooth"], lw=2.0, label="smoothed")
    plt.axhline(0.0, ls="--", lw=1.0)

    if rankcorr["k_eff_0_raw"] is not None and np.isfinite(rankcorr["k_eff_0_raw"]):
        plt.axvline(rankcorr["k_eff_0_raw"], ls=":", lw=1.2, label=f"k_eff_0_raw={rankcorr['k_eff_0_raw']:.1f}")
    if rankcorr["k_eff_0_smooth"] is not None and np.isfinite(rankcorr["k_eff_0_smooth"]):
        plt.axvline(rankcorr["k_eff_0_smooth"], ls="-.", lw=1.2, label=f"k_eff_0_smooth={rankcorr['k_eff_0_smooth']:.1f}")
    if rankcorr["k_eff_50"] is not None and np.isfinite(rankcorr["k_eff_50"]):
        plt.axvline(rankcorr["k_eff_50"], ls="--", lw=1.0, alpha=0.8, label=f"k_eff_50={rankcorr['k_eff_50']:.0f}")
    if rankcorr["k_eff_90"] is not None and np.isfinite(rankcorr["k_eff_90"]):
        plt.axvline(rankcorr["k_eff_90"], ls="--", lw=1.0, alpha=0.8, label=f"k_eff_90={rankcorr['k_eff_90']:.0f}")

    plt.xlabel("topological rank n")
    plt.ylabel("rank correlation C^(n)")
    plt.title("Empirical topological-rank correlation")
    plt.legend(fontsize=8)
    savefig(outdir / "rank_correlation.png", dpi=dpi)


def plot_proxy_vs_k(proxy_df: pd.DataFrame, cfg: EmpiricalConfig, outdir: Path, dpi: int) -> None:
    fig, ax = plt.subplots(figsize=(6.4, 4.6))
    ax.plot(proxy_df["k"], proxy_df["c_mis_mean"], marker="o", lw=1.5, label="c_mis(k)")
    for k in cfg.core_window_k:
        ax.axvspan(k - 0.35, k + 0.35, alpha=0.08)
    ax.set_xlabel("k used in empirical local neighborhood")
    ax.set_ylabel("local-misalignment cost proxy")
    ax.set_title("Empirical local cost proxy vs k")

    if "J_emp_proxy" in proxy_df.columns:
        ax2 = ax.twinx()
        ax2.plot(proxy_df["k"], proxy_df["J_emp_proxy"], marker="s", lw=1.5, label="J_emp_proxy(k)")
        ax2.set_ylabel("empirical response-cost proxy")
        h1, l1 = ax.get_legend_handles_labels()
        h2, l2 = ax2.get_legend_handles_labels()
        ax.legend(h1 + h2, l1 + l2, loc="best")
    else:
        ax.legend(loc="best")

    savefig(outdir / "proxy_vs_k.png", dpi=dpi)


def plot_xy_velocity_snapshot(r1: jnp.ndarray, u: jnp.ndarray, outdir: Path, dpi: int) -> None:
    r1_np = np.asarray(r1)
    u_np = np.asarray(u)
    plt.figure(figsize=(6.4, 5.0))
    plt.quiver(
        r1_np[:, 0], r1_np[:, 1],
        u_np[:, 0], u_np[:, 1],
        angles="xy", scale_units="xy", scale=1.0, width=0.002,
    )
    plt.xlabel("x(t1)")
    plt.ylabel("y(t1)")
    plt.title("2D projection of empirical velocity field")
    savefig(outdir / "velocity_field_xy.png", dpi=dpi)


def plot_regime_overlay(
    model_df: Optional[pd.DataFrame],
    phi_emp: float,
    xi_emp: float,
    outdir: Path,
    dpi: int,
) -> None:
    if model_df is None:
        return

    work = model_df.copy()
    phi_col = None
    xi_col = None
    color_col = None
    for cand in ["phi_mean", "phi", "Phi"]:
        if cand in work.columns:
            phi_col = cand
            break
    for cand in ["xi", "xi_mean", "corr_length"]:
        if cand in work.columns:
            xi_col = cand
            break
    for cand in ["J_min", "J", "J_mean"]:
        if cand in work.columns:
            color_col = cand
            break
    if phi_col is None or xi_col is None:
        return

    plt.figure(figsize=(6.2, 4.8))
    if color_col is not None:
        sc = plt.scatter(work[phi_col], work[xi_col], c=work[color_col], s=24)
        plt.colorbar(sc, label=color_col)
    else:
        plt.scatter(work[phi_col], work[xi_col], s=24, alpha=0.7)
    plt.scatter([phi_emp], [xi_emp], marker="*", s=220, edgecolors="k", linewidths=0.8, label="empirical event")
    plt.xlabel(phi_col)
    plt.ylabel(xi_col)
    plt.title("Empirical event overlaid on model regime map")
    plt.legend()
    savefig(outdir / "regime_overlay.png", dpi=dpi)


# -----------------------------
# 9. Reporting / serialization
# -----------------------------

def write_json(path: str | Path, obj: Dict[str, Any]) -> None:
    with open(path, "w", encoding="utf-8") as f:
        json.dump(obj, f, indent=2, ensure_ascii=False)


def summarize_results(
    cfg: EmpiricalConfig,
    basic: Dict[str, Any],
    metric: Dict[str, Any],
    rankcorr: Dict[str, Any],
    proxy_df: pd.DataFrame,
    nearest_rows: Optional[pd.DataFrame],
) -> Dict[str, Any]:
    summary = {
        "input_path": cfg.input_path,
        "delta_t": cfg.delta_t,
        "n_birds": int(basic["r1"].shape[0]),
        "phi_emp": float(basic["phi"]),
        "mean_speed": float(basic["mean_speed"]),
        "xi_emp_zero_cross": None if not np.isfinite(metric["xi_0"]) else float(metric["xi_0"]),
        "xi_emp_integral": None if not np.isfinite(metric["xi_int"]) else float(metric["xi_int"]),
        "chi_C_emp": None if not np.isfinite(metric["chi_C"]) else float(metric["chi_C"]),
        "k_eff_0_raw": None if rankcorr["k_eff_0_raw"] is None or not np.isfinite(rankcorr["k_eff_0_raw"]) else float(rankcorr["k_eff_0_raw"]),
        "k_eff_0_smooth": None if rankcorr["k_eff_0_smooth"] is None or not np.isfinite(rankcorr["k_eff_0_smooth"]) else float(rankcorr["k_eff_0_smooth"]),
        "k_eff_50": None if rankcorr["k_eff_50"] is None or not np.isfinite(rankcorr["k_eff_50"]) else float(rankcorr["k_eff_50"]),
        "k_eff_75": None if rankcorr["k_eff_75"] is None or not np.isfinite(rankcorr["k_eff_75"]) else float(rankcorr["k_eff_75"]),
        "k_eff_90": None if rankcorr["k_eff_90"] is None or not np.isfinite(rankcorr["k_eff_90"]) else float(rankcorr["k_eff_90"]),
        "core_window_k": list(cfg.core_window_k),
        "shoulder_k": list(cfg.shoulder_k),
        "proxy_table": proxy_df.to_dict(orient="records"),
    }
    if nearest_rows is not None:
        summary["nearest_model_rows"] = nearest_rows.to_dict(orient="records")
    return summary


# -----------------------------
# 10. Main pipeline
# -----------------------------

def run_empirical_anchor(cfg: EmpiricalConfig) -> Dict[str, Any]:
    outdir = Path(cfg.outdir)
    outdir.mkdir(parents=True, exist_ok=True)

    df = parse_sd01_ascii_table(cfg.input_path)
    print(f"Parsed dataset shape: {df.shape}")
    df.to_csv(outdir / "parsed_positions.csv", index=False)

    basic = compute_basic_state(df, cfg.delta_t)

    bird_df = df.copy()
    bird_df[["vx", "vy", "vz"]] = np.asarray(basic["v"])
    bird_df[["ux", "uy", "uz"]] = np.asarray(basic["u"])
    bird_df[["dux", "duy", "duz"]] = np.asarray(basic["delta_u"])
    bird_df["speed"] = np.asarray(basic["speed"])
    bird_df.to_csv(outdir / "bird_level_measurements.csv", index=False)

    metric = metric_connected_correlation(
        r=basic["r1"],
        delta_u=basic["delta_u"],
        n_bins=cfg.n_r_bins,
        r_max_quantile=cfg.r_max_quantile,
        min_pairs_per_bin=cfg.min_pairs_per_bin,
    )
    metric_df = pd.DataFrame({
        "r_center": metric["r_centers"],
        "C_r": metric["C_r"],
        "pair_count": metric["counts"],
    })
    metric_df.to_csv(outdir / "metric_correlation.csv", index=False)

    rankcorr = topological_rank_correlation(
        r=basic["r1"],
        delta_u=basic["delta_u"],
        max_rank=cfg.max_rank,
        smooth_window=cfg.rank_smooth_window,
    )
    rank_df = pd.DataFrame({
        "rank": rankcorr["ranks"],
        "C_n": rankcorr["C_n"],
        "C_n_smooth": rankcorr["C_n_smooth"],
    })
    rank_df.to_csv(outdir / "topological_rank_correlation.csv", index=False)

    mismatch_df = empirical_local_mismatch_vs_k(
        u=basic["u"],
        r=basic["r1"],
        k_grid=cfg.k_grid,
    )
    proxy_df = build_empirical_proxy_table(mismatch_df, metric["chi_C"])
    proxy_df.to_csv(outdir / "empirical_proxy_scan.csv", index=False)

    model_df = maybe_load_csv(cfg.model_window_csv)
    nearest_rows = None
    if model_df is not None and np.isfinite(metric["xi_0"]):
        nearest_rows = nearest_model_rows(model_df, basic["phi"], metric["xi_0"], n_select=5)
        if nearest_rows is not None:
            nearest_rows.to_csv(outdir / "nearest_model_rows.csv", index=False)

    if cfg.save_plots:
        plot_xy_velocity_snapshot(basic["r1"], basic["u"], outdir, cfg.dpi)
        plot_metric_correlation(metric, outdir, cfg.dpi)
        plot_rank_correlation(rankcorr, outdir, cfg.dpi)
        plot_proxy_vs_k(proxy_df, cfg, outdir, cfg.dpi)
        plot_regime_overlay(model_df, basic["phi"], metric["xi_0"], outdir, cfg.dpi)

    summary = summarize_results(cfg, basic, metric, rankcorr, proxy_df, nearest_rows)
    write_json(outdir / "summary.json", summary)

    return {
        "summary": summary,
        "bird_df": bird_df,
        "metric_df": metric_df,
        "rank_df": rank_df,
        "proxy_df": proxy_df,
        "nearest_rows": nearest_rows,
    }


# -----------------------------
# 11. Example usage in notebook
# -----------------------------

cfg = EmpiricalConfig(
    input_path="sd01.rtf",
    outdir="./exp6_empirical_anchor_v2_sd01",
    delta_t=1.0,
    n_r_bins=48,
    r_max_quantile=0.95,
    min_pairs_per_bin=50,
    max_rank=100,
    rank_smooth_window=5,
    k_grid=(3, 4, 5, 6, 7, 9, 12, 15, 20),
    core_window_k=(3, 4, 5),
    shoulder_k=(6, 7),
    model_window_csv=None,   # optional: e.g. "exp4_window_summary.csv"
    save_plots=True,
    dpi=180,
)

results = run_empirical_anchor(cfg)

print(json.dumps(results["summary"], indent=2, ensure_ascii=False))

results["metric_df"].head(), results["rank_df"].head(), results["proxy_df"].head()
```
